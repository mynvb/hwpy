```python
#1a
def prob_count(n=3):
    '''Событие А = Ладьи не бьют друг друга
    Благоприятные исходы: когда поставили ладью на доску (одна итерация), исключаем ряд и колонку из кол-ва исходов (например, если доска 4 на 4, после первой ладьи уже количество вариантов 3 на 3)
    Все исходы: при каждой итерации количество клетов на доске уменьшается на 1, так как ладью не убрали
    Количество исходов на каждой итерации перемножается между собой после каждой итерации '''
    outcomes, favor_outcomes = 1, 1
    for i in range(n):
        outcomes = outcomes * ((n*n)-i)
        favor_outcomes = favor_outcomes * (n-i)*(n-i)
    
    return favor_outcomes / outcomes

prob_count(n=5)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
x = np.linspace(0,30,31)
#графики плотности
for mu in [(1,'bo'),(4,'go'),(15, 'ro')]:
    poisson_cdf = sts.poisson(mu[0]).cdf(x)
    plt.plot(x, poisson_cdf, mu[1], ms=5, label = "$mu=%s$" % (mu[0]))
plt.legend()
plt.title("CDF (Poisson)")
plt.ylabel('$p(x)$')
plt.xlabel('$x$')
```

```python
x = np.linspace(0,30,31)
for mu in [(1,'bo'),(4,'go'),(15, 'ro')]:
    poisson_pmf = sts.poisson(mu[0]).pmf(x)
    plt.plot(x, poisson_pmf, mu[1], ms=5, label = "$mu=%s$" % (mu[0]))
plt.legend()
plt.title("PMF (Poisson)")
plt.ylabel('$p(x)$')
plt.xlabel('$x$')
```

```python
x = np.linspace(0,10,100)

for lambd in [1,3,6]:
    exp_cdf = sts.expon(lambd).cdf(x)
    plt.plot(x, exp_cdf, label = "$l=%s$" % (lambd))
plt.legend()
plt.title("CDF (Exponential)")
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
for lambd in [1,3,6]:
    exp_pmf = sts.expon(lambd).pdf(x)
    plt.plot(x, exp_pmf, label = "$l=%s$" % (lambd))
plt.legend()
plt.title("PDF (Exponential)")
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
x = np.linspace(0,100,100)
for a in [20,55,100]:
    gamma_cdf = sts.gamma(a).cdf(x)
    plt.plot(x, gamma_cdf, label = "$a=%s$" % (a))
plt.legend()
plt.title("CDF (Gamma)")
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
for a in [20,55,100]:
    gamma_pdf = sts.gamma(a).pdf(x)
    plt.plot(x, gamma_pdf, label = "$a=%s$" % (a))
plt.legend()
plt.title("PDF (Gamma)")
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```
