```python
#1a
def prob_count(n=3):
    'Событие А = Ладьи не бьют друг друга'
    outcomes, favor_outcomes = 1, 1
    for i in range(n):
        outcomes = outcomes * ((n*n)-i)
        favor_outcomes = favor_outcomes * (n-i)*(n-i)
    
    return favor_outcomes / outcomes

prob_count(n=6)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline

```

```python
x = np.linspace(0,30,31)
poisson1 = sts.poisson(1)
poisson2 = sts.poisson(4)
poisson3 = sts.poisson(15)
```

```python
#графики плотности
cdf1 = poisson1.cdf(x)
cdf2 = poisson2.cdf(x)
cdf3 = poisson3.cdf(x)
plt.step(x, cdf1)
plt.step(x, cdf2)
plt.step(x, cdf3)

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
pmf1 = poisson1.pmf(x)
pmf2 = poisson2.pmf(x)
pmf3 = poisson3.pmf(x)

plt.plot(x, pmf1)
plt.plot(x, pmf2)
plt.plot(x, pmf3)

plt.ylabel('$P(X=x)$')
plt.xlabel('$x$')
```

```python
x = np.linspace(0,10,100)
exp1 = sts.expon(1)
exp2 = sts.expon(3)
exp3 = sts.expon(6)

cdf1 = exp1.cdf(x)
cdf2 = exp2.cdf(x)
cdf3 = exp3.cdf(x)

plt.plot(x, cdf1)
plt.plot(x, cdf2)
plt.plot(x, cdf3)

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
pdf1 = exp1.pdf(x)
pdf2 = exp2.pdf(x)
pdf3 = exp3.pdf(x)

plt.plot(x, pdf1)
plt.plot(x, pdf2)
plt.plot(x, pdf3)
```

```python
x = np.linspace(0,100,100)
g1 = sts.gamma(20)
g2 = sts.gamma(100)
g3 = sts.gamma(55)

cdf1 = g1.cdf(x)
cdf2 = g2.cdf(x)
cdf3 = g3.cdf(x)

plt.plot(x, cdf1)
plt.plot(x, cdf2)
plt.plot(x, cdf3)

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
pdf1 = g1.pdf(x)
pdf2 = g2.pdf(x)
pdf3 = g3.pdf(x)

plt.plot(x, pdf1)
plt.plot(x, pdf2)
plt.plot(x, pdf3)
```
